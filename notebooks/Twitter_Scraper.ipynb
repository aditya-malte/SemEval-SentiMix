```
keyword_sets =  ["diwali OR eid OR ganapati OR holi OR new year",
                  "modi OR yogi OR kejriwal OR arnab OR rahul" ,
                  "thandi OR garmi OR baarish OR tufaan" ,
                  "aadmi OR sirji OR bhaiyya OR behen" ,
                  "amitabh OR shahrukh OR amir OR deepika" ,
                  "mumbai OR delhi OR chennai OR kolkata OR pune OR bangalore",
                  "chandrayaan OR mangalyaan OR nasa OR isro",
                  "sindhu OR kohli OR cricket OR ipl",
                  "india OR pakistan OR usa OR china OR bangladesh",
                  "pokhran OR pulwama OR kargil OR kashmir",
                  "jio OR vodafone OR idea OR airtel",
                  "hindu OR sikh OR muslim OR jain",
                  "army OR airforce OR navy OR javaan",
                  "gaadi OR bangla OR ghar OR bike",
                  "gai OR bakri OR sher OR gadha OR kutta OR pakshi",
                  ]

print(keyword_sets)
```
